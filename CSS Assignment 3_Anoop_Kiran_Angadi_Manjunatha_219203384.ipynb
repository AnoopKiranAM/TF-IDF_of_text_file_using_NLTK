```python
from nltk.tokenize import sent_tokenize,word_tokenize
file_content = open(r"Article.txt", encoding="utf-8").read()
sentences=sent_tokenize(file_content)
total_sentense_length=len(sentences)
print("Total Number of sentences:",total_sentense_length)
all_sentences = [] 
for i in sentences:
    all_sentences.append(i.replace("-"," "))
```

```python
def total_words_count(str):
    words=word_tokenize(str)
    return words
```

```python
unique_words = set()
for i in all_sentences:
    words=total_words_count(i)
    unique_words.update(words)
print(len(unique_words))
print("Unique Words=",unique_words)
```

```python
def word_count(str):
    word_frequency_in_sentence = dict()
    words=word_tokenize(str) #tokenize the string into words using word_tokenize
    for word in words:
        if word.lower() in word_frequency_in_sentence:  #Converting everything to lower to avoide duplicate entries
            word_frequency_in_sentence[word.lower()] += 1
        else:
            word_frequency_in_sentence[word.lower()] = 1

    return word_frequency_in_sentence
```

```python
def word_frequency(str):
    term_frequency_each_sentence = dict()
    words=word_tokenize(str) #tokenize the string into words using word_tokenize
    for word in words:
        if word.lower() in term_frequency_each_sentence:  #Converting everything to lower to avoide duplicate entries
            term_frequency_each_sentence[word.lower()] += 1/len(words)
        else:
            term_frequency_each_sentence[word.lower()] = 1/len(words)

    return term_frequency_each_sentence
```

```python
def word_in_sentences(termFrequency_Matrix):
    word_per_doc_table = {}

    for sent, value in termFrequency_Matrix.items():
        for word, count in value.items():
            if word.lower() in word_per_doc_table:
                word_per_doc_table[word.lower()] += 1
            else:
                word_per_doc_table[word.lower()] = 1

    return word_per_doc_table
```

```python
import math
def generate_idf_matrix(termFrequency_Matrix, sentences_having_each_word):
    idf_matrix = {}
    for sent, value in termFrequency_Matrix.items():
        idf_table = {}
        for word in value.keys():
            idf_table[word] = math.log10(total_sentense_length / float(sentences_having_each_word[word]))
        idf_matrix[sent] = idf_table
    return idf_matrix
```

```python
def generate_tf_idf_matrix(tf_matrix, idf_matrix):
    tf_idf_matrix = {}
    for (sent1, f_table1), (sent2, f_table2) in zip(tf_matrix.items(), idf_matrix.items()):
        tf_idf_table = {}
        for (word1, value1), (word2, value2) in zip(f_table1.items(),f_table2.items()):
            tf_idf_table[word2] = float(value1 * value2)
        tf_idf_matrix[sent1] = tf_idf_table
    return tf_idf_matrix
```

```python
sentDict_with_wordFrequency = dict.fromkeys(all_sentences, 0)
termFrequency_Matrix = dict.fromkeys(all_sentences, 0)
for i in all_sentences:
    sentDict_with_wordFrequency[i]=word_count(i) #word frequency in each sentense
    termFrequency_Matrix[i]=word_frequency(i)   #term frequency matrix in each sentense
```

```python
def adding_weight_to_each_sentences(tf_idf_matrix):
    sentenceValue = {}
    for sent, value in tf_idf_matrix.items():
        total_score_per_sentence = 0
        count_words_in_sentence = len(value)
        for word, score in value.items():
            total_score_per_sentence += score
        sentenceValue[sent] = total_score_per_sentence / count_words_in_sentence
    return sentenceValue
```

```python
def average_sentence_weight(weighted_sentence):
    value = 0
    for k in weighted_sentence:
        value += weighted_sentence[k]
    average = (value / len(weighted_sentence))
    return average
```
